```python
# Pivot Tables = allowing for the transformation of raw data into a more concise and insightful summary.



df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})

df
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc="sum",fill_value = 0)
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                       aggfunc={'D': "mean", 'E': "mean"})
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                       aggfunc={'D': "mean",
                                'E': ["min", "max", "mean"]})
table
```

```python
#columns
table =pd.pivot_table(df,index=['D'],columns=['E'],values=['C'],aggfunc=np.sum,fill_value=0)
table
```

```python
#  Exercise 15:
#  Group data by region and category.
#  Calculate total sales and average order value.
#  Sort regions by total revenue.
```

```python
group_data = {
    'region': ['Asia', 'Asia', 'Asia', 'Asia', 'Europe', 'Europe', 'Europe', 'North America', 'North America', 'North America'],
    'category' : ['China', 'India', 'Japan', 'Nepal', 'France', 'Germany', 'United Kingdom', 'Canada', 'Mexico', 'United States'],
    'sales': [20000, 40000, 50000, 60000,40000, 70000, 20000, 10000, 90000,30000]
}

df = pd.DataFrame(group_data)
df


grouped_data = df.groupby('region')['category'].sum()
grouped_data
```

```python
#  Calculate total sales and average order value.

total_sales = df.groupby('region')['sales'].sum()
print(total_sales)

total_sales_by_category = df.groupby('category')['sales'].sum()
print(total_sales_by_category)
```

```python
df['total_sales_by_category'] = df.groupby('category')['sales'].transform('sum')
df

```

```python
df['total_sales_by_region'] = df.groupby('region')['sales'].transform('sum')
df
```

```python
# Sort regions by total revenue.

sorted_values =  df.sort_values( by = 'total_sales_by_region',ascending=False)
sorted_values
```

```python
# # Use rolling averages to smooth out daily sales data over 7-day windows

rolling_sales_data = {
    'dailysales':['Monday','Tuesday','Wednesday','Thursday', 'Friday','Saturday','Sunday','Monnday','Tuesday','Wednesday','Thursday','Friday','Saturday','sunday'],
     'salescount':[20000,10000,30000,40000,70000,60000,30000,20000,50000,30000,11000,70000,10000,30000]
}

sales = pd.DataFrame(rolling_sales_data)
sales


sales['7_day_rolling_avg'] = sales['salescount'].rolling(window=1).mean()
sales


```
